```python
import pandas as pd
import numpy as np
df=pd.read_csv('laptop_data(1).csv')
df
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
for column in df.columns:
    print(df[column].value_counts())
    print("*"*50)
```

```python
df.drop(columns=['Unnamed: 0','ScreenResolution','Gpu','Weight'],inplace = True)
```

```python
df['TypeName'] = df['TypeName'].replace('2 in 1 Convertible', 'Two in One Convertible')
```

```python
df['GB'] = df['Ram'].str.extract('(\d+)').astype(int)
```

```python
df['Size'] = df['Memory'].str.extract('(\d+)').astype(int)
df.head()
```

```python
df['CPU'] = df['Cpu'].str.split().str.slice(0, 3).str.join(' ')
df.head()
```

```python
df.drop(columns=['Cpu','Ram','Memory'],inplace = True)
```

```python
df
```

```python
x=df[['Company','TypeName','Inches','OpSys','GB','Size','CPU']]
y=df[['Price']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
ohe=OneHotEncoder()
ohe.fit(x[['Company','TypeName','OpSys','CPU']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['Company','TypeName','OpSys','CPU']), remainder='passthrough')
```

```python
scaler=StandardScaler()
scaler = StandardScaler(with_mean=False)
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred_lr)
```

```python
lasso=Lasso()
pipe=make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lasso=pipe.predict(x_test)
r2_score(y_test,y_pred_lasso)
```

```python
#Ridge
```

```python
ridge=Ridge()
pipe=make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_ridge=pipe.predict(x_test)
r2_score(y_test,y_pred_ridge)
```

```python
print("No regularization:",r2_score(y_test,y_pred_lr))
print("Lasso:",r2_score(y_test,y_pred_lasso))
print("Ridge:",r2_score(y_test,y_pred_ridge))
```
